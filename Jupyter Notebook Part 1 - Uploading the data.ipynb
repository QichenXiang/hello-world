```python
##############################################################
#CREATE TABLE 'Serious_Incidents'
##############################################################


cursor = connection.cursor()
db_name = 'philadelphia'
table_name = 'Serious_Incidents'
# Create a table
# The {0} and {1} are placeholders for the parameters in the format(....) statement
create_table_query = '''CREATE TABLE IF NOT EXISTS {0}.{1} 
 (ULCS_CODE INT(4) NULL,
  SRC_ID VARCHAR(45) NOT NULL,
  Incident_Type VARCHAR(45) NULL,
  Incident_Count INT NULL
   )'''.format(db_name, table_name)
cursor.execute(create_table_query)
cursor.close()
```

```python
#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
#UPLOAD DATA to 'Serious_Incidents'
#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!


#open the csv file, pass the content to variable 'data'
table=open('Table 2 - Serious Incidents 2016-2017.csv','r')
data=table.read().split('\n')[1:]

#import the data into MYSQL server
query_template = '''
INSERT INTO 
philadelphia.Serious_Incidents(
  ULCS_CODE,
  SRC_ID,
  Incident_Type,
  Incident_Count
)
        VALUES (%s, %s, %s, %s)'''

cursor = connection.cursor()

for entry in data:
    entry=entry.split(',')
    ULCS_CODE = entry[0]
    SRC_ID = entry[1]
    Incident_Type = entry[2]
    Incident_Count = entry[3]    
    query_parameters = (ULCS_CODE, SRC_ID,Incident_Type, Incident_Count)
    
    cursor.execute(query_template, query_parameters)
'''
Commit changes to stable storage
'''
connection.commit()
cursor.close()
```
